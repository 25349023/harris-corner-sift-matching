```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
def normalize(data, lo, hi):
    return cv2.normalize(data, None, lo, hi, norm_type=cv2.NORM_MINMAX)
```

```python
def imshow(image):
    image = normalize(image, 0, 255)
    plt.imshow(cv2.cvtColor(image.astype(np.uint8), cv2.COLOR_BGR2RGB))
```

```python
img_left = cv2.imread('1a_notredame.jpg')
imshow(img_left)
```

```python
gray = cv2.cvtColor(img_left, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create(200)
kp, descriptor = sift.detectAndCompute(gray, None)
keypoints = cv2.drawKeypoints(gray, kp, img_left)
```

```python
points = np.array([k.pt for k in kp])
points
```

```python
plt.figure(figsize=(10, 15))
plt.imshow(img_left)
plt.scatter(points[:, 0], points[:, 1], c='r', s=5)
```
